```python
#função para inserir uma coluna com flag  que diz se o investimento 
#é maior que 100mil

def maior_100k(valor):
    if valor > 1000000:
        return 1
    else:
        return 0
```

```python
#criação de nova coluna 'maior_100k'
investimentos['maior_100k'] = investimentos['valor'].apply(maior_100k)
```
