```python
import pandas as pd
import psycopg2
from sqlalchemy import create_engine
```

```python
conn = create_engine('postgresql://postgres:151322@localhost:5432/postgres')
```

```python
query = 'SELECT * FROM investimento'
investimentos = pd.read_sql(query, conn)
```

```python
query2 = 'SELECT * FROM investidor'
investidores = pd.read_sql(query2, conn)
```

```python
query3 = 'SELECT * FROM startup'
startups = pd.read_sql(query3, conn)
```

```python
investimentos.head()
```

```python
# metodo merge() que faz o msm papel do JOIN no sql
# nesse caso trazer o nome dos investidores para a juntar com a tabela  de investimentos
final1 = investimentos.merge(investidores, how='inner', left_on='investidor_id', right_on='id')
```

```python
final2 = investimentos.merge(startups, how='inner', left_on='startup_id', right_on='id')
```

### 1. A tabela abaixo apresenta o faturamento de uma empresa nos primeiro trimestre do ano de 2018:

Mês | Faturamento
:---|:----------- 
Janeiro | 420000
Fevereiro | 260000
Março | 310000

### A tabela a seguir, por outro lado, apresenta o faturamento dessa mesma empresa no primeiro trimestre do ano de 2019.

Mês | Faturamento
:---|:----------- 
Janeiro | 290000
Fevereiro | 275000
Março | 442000

### Com base nesses dados, crie dois DataFrames com os dados dessas duas tabelas. Em seguida, utilize o método merge para juntar esses dois DataFrames, de modo que o novo DataFrame tenha apenas três colunas: o mês, o faturamento em 2018 e o faturamento em 2019.

```python
dados2018 ={
    'Mês': ['Janeiro', 'Fevereiro', 'Março'],
    'Faturamento 2018': [420000, 260000, 310000]
}

dados2019 ={
    'Mês': ['Janeiro', 'Fevereiro', 'Março'],
    'Faturamento 2019': [290000, 275000, 442000]
}
```

```python
df1 = pd.DataFrame(dados2018)
df1
```

```python
df2 = pd.DataFrame(dados2019)
df2
```

```python
final = df1.merge(df2, how='inner', on='Mês')
final.head()
```
