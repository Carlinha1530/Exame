```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    'A': [5, 2, np.nan],
    'B': [np.nan, 2, np.nan],
    'C': ['b', 'a', '']
})

df
```

## O metodo .insnull() verificar os valores nulos

```python
#Verfica se a coluna A tem valores nulos
df['A'].isnull()
```

```python
#Para pegar apenas as colunas que possuem dados é preciso negas a mascara anterior
df[~df['A'].isnull()]
```

## O metodo .dopna() exclui valores nan(nulos)
## Excluir linhas que possuem ao menos um elemneto nulo

```python
df.dropna()
```

## Com os metodos
## .axis() que pode sobrescrever comportamento de exclusão de linhas. Torna exclusão por colunas

## ou
## .thresh() que um metodo de limite, pode especificar a partir de quantos nulos se exclui a linha ou coluna

```python
df.dropna(1)
```

```python
df.dropna(thresh=2)
```

```python
df.dropna(axis=1, thresh=2)
```

## O método .fillna() informa valor para substituir cada um dos dados faltantes

```python
#preenche os valores faltantes com valor informado
df.fillna(1)
```

```python
# preeenche valor faltante com o proximo valor não nulo. Utilizando o metodo ffill
df.fillna(method='ffill')
```

```python
# preeenche valor faltante com o ultimo valor não nulo. Utilizando o metodo bfill
df.fillna(method='bfill')
```

```python
# preeenche valor faltante com um valor qualquer que será buscado no dataframe e o substitui com um outro informado.
df.replace(to_replace=np.nan, value=1)
```
